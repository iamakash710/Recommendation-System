## Data mining

```python
user_book_df = data.pivot_table(index = 'user_id',
                                 columns = 'book_title',
                                 values = 'book_rating').reset_index(drop=True)

user_book_df
```

```python
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

## checking the user similarity using cosine metric

```python

us_sim = 1-pairwise_distances(user_book_df, metric='cosine' )

#storing the values in a dataframe
us_sim_df=pd.DataFrame(us_sim)

#converting the user indexes into the user id
us_sim_df.index=data.user_id.unique()
us_sim_df.columns=data.user_id.unique()

us_sim_df.head()
```

```python
us_sim_df.shape
```

- We can see that the diagonal values are 1, it may cause error while performing the recommendation , we will remove them.

```python
np.fill_diagonal(us_sim, 0)
us_sim_df.head()
```

- Now we have all pure simillarity values

```python
# now we can see which users are correlated.
us_sim_df.idxmax(axis=1)
```
